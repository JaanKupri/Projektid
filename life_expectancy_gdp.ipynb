Goals:
1) Visualize the relationship between year and life expectancy
2) Visualize the relationship between country and life expectancy
3) Does a country's life expenctancy grow alongside GDP?


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('all_data.csv', encoding='utf-8')

print(df.head())
```

```python
min_life_expectancy_idx = df['Life expectancy at birth (years)'].idxmin()
print("Lowest life expectancy:")
print(df.loc[[min_life_expectancy_idx]])
print(" ")
max_life_expectancy_idx = df['Life expectancy at birth (years)'].idxmax()
print("Highest life expectancy:")
print(df.loc[[max_life_expectancy_idx]])
print(" ")
sum = 0
for i in df['Life expectancy at birth (years)']:
  sum += i

print("Average life expectancy:")
print(sum / len(df))
```

```python
unique_countries = df.Country.unique()

for country in unique_countries:
  plt.figure(figsize=(10, 6))
  country_data = df[df['Country'] == country]

  plt.bar(country_data['Year'], country_data['Life expectancy at birth (years)'])
  if country == 'Zimbabwe':
    plt.axis([1999, 2016, 40, 65])
  else:
    plt.axis([1999, 2016, 70, 82])
  plt.title(f'Life Expectancy for {country}')
  plt.xlabel('Year')
  plt.ylabel('Life Expectancy')

  plt.show()
  plt.clf()
```

```python
for country in unique_countries:
  plt.figure(figsize=(10, 6))
  country_data = df[df['Country'] == country]

  plt.bar(country_data['Year'], country_data['GDP'])

  plt.title(f'GDP {country}')
  plt.xlabel('Year')
  plt.ylabel('GDP')

  plt.show()
  plt.clf()
```

```python
for country in unique_countries:
  plt.figure(figsize=(10,6))
  country_data = df[df['Country'] == country]
  plt.scatter(country_data['Life expectancy at birth (years)'], country_data['GDP'])
  plt.xlabel('Life expectancy')
  plt.ylabel('GDP')
  plt.show()
  plt.clf()
```

```python
for country in unique_countries:
  country_data = df[df['Country'] == country]
  sum = 0
  country_mean = country_data['Life expectancy at birth (years)'].mean()
  plt.bar(country, country_mean)
plt.axhline(y=df['Life expectancy at birth (years)'].mean(),linewidth=1, color='k', label='average')
plt.xticks(rotation=45, fontsize=8)
plt.axis([-0.5, 5.5, 45, 81])
plt.title("Average life expectancy per country (Horizontal line representing the average)")
plt.show()
plt.clf()

```
